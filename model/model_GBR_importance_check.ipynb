# Cleaning data

```python
import pandas as pd
import numpy as np
```

```python
df_airbnb_cleaned = df_airbnb.drop(["title","url"], axis=1).dropna().reset_index(drop=True)
df_airbnb_cleaned
```

```python
np.sort(df_airbnb_cleaned["property_type"].unique())
```

```python
df_airbnb_cleaned = df_airbnb_cleaned.drop([1555,1355], axis=0).reset_index(drop=True)
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Encode cities and property categorical features
enc = OneHotEncoder()
transformed = enc.fit_transform(df_airbnb_cleaned[["city","property_type","location"]]).toarray()
df_enc = pd.DataFrame(transformed, columns=enc.get_feature_names(["city","type","location"]))
df_enc
```

```python
# Join main airbnb dataframe with encoded dataframe features
df_airbnb_enc = pd.concat([df_airbnb_cleaned, df_enc], axis=1).drop(["city","property_type","location"], axis=1)
df_airbnb_enc
```

```python
# Check features and target correlation
import seaborn as sns
import matplotlib.pyplot as plt

plt.subplots(figsize=(50,20))
corr = df_airbnb_enc.corr(method="pearson")
sns.heatmap(corr)
```

# Modeling


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error
```

```python
# Creating train and test dataframes with 80% and 20% ratio.
df_test = pd.DataFrame()
df_train = pd.DataFrame()
for city in df_airbnb["city"].unique():  
  boolean = df_airbnb_enc[f"city_{city}"] == 1
  df_city = df_airbnb_enc[boolean]
  df_city_train, df_city_test = np.split(df_city, [int(.8*len(df_city))])
  df_train = pd.concat([df_train, df_city_train])
  df_test = pd.concat([df_test, df_city_test])
```

```python
 df_train
```

```python
len(df_train), len(df_test)
```

```python
X_train = df_train.drop(columns=["price"])
X_test = df_test.drop(columns=["price"])
y_train = df_train["price"]
y_test = df_test["price"] 
```

```python
reg = GradientBoostingRegressor()
reg.fit(X_train, y_train)

predicted = reg.predict(X_test)
```

```python
def feat_importance(m, df):
    return pd.DataFrame({'cols':df.columns, 'imp':m.feature_importances_})
```

```python
rf_fi = feat_importance(reg, X_train)
rf_fi = rf_fi.sort_values(ascending=False, by='imp').reset_index(drop=True)
```

```python
plt.figure(figsize=(12, 40))
ax = sns.barplot(x="imp", y="cols", data=rf_fi)
```

```python
# Get the coefficient of determination R squared. Check how well given target can be predicted using a set of features.
reg.score(X_test,y_test)
```

```python
# Mean absolute error
mean_absolute_error(y_test, predicted)
```
